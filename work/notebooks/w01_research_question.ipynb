What decision does your work improve? The ordering/sorting of flight options presented to a user during a search session.

Who acts on it? The front-end search engine results page (SERP) ranking service, which will dynamically arrange flight offers based on the model's scores.

What does a wrong recommendation cost? A bad ranking causes immediate user friction. If relevant or competitively priced flights are buried on page 2 or lower down the list, the user is highly likely to bounce to a competitor site (e.g., Google Flights or Expedia), resulting in lost click revenue and missed booking commissions

```python
# A quick mock calculation of cost impact if bounce rate increases
assumed_monthly_searches = 100000
current_conversion = 0.04
commission_per_booking = 15

# If a bad ranking causes a 10% drop in conversion
lost_revenue = assumed_monthly_searches * (current_conversion * 0.10) * commission_per_booking
print(f"Estimated monthly risk exposure of poor ranking: ${lost_revenue:,}")
```
